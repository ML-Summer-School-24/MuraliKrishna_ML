```python

# 1.Importing Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import accuracy_score
from colorama import Fore,Style
from sklearn.ensemble import RandomForestClassifier

print(Fore.GREEN + Style.BRIGHT + 'THE REQUIRED LIBRARIES WERE SUCCESSFULLY IMPORTED')
```

```python

# 2.Loading the CSV file

filePath = r"C:\Users\Murali Krishna\Downloads\winequality-red.csv"
data = pd.read_csv(filePath)
data
```

```python

# 3.Data Pre-Processing


# 3.1 Deleting the outliers
print("Before Removing the outliers", data.shape)
data = data[data['total sulfur dioxide']<160]
print("After Removing the outliers", data.shape)
print("-----------------------------------------------------------------")

X = data.drop('quality',axis=1)
Y = data['quality'].apply(lambda y_value: 1 if y_value >= 5 else 0)
print("Shape of X:", X.shape)
print("Shape of Y:", Y.shape)
print("-----------------------------------------------------------------")

# 3.2 Handling Missing Values
null_counts = data.isnull().sum()
print(null_counts)
print(Fore.GREEN + Style.BRIGHT + "Totally, there are {null_counts.sum()} null values in the dataset.")
print("-----------------------------------------------------------------")

# 3.3 Handling Duplicates
duplicate_count = data.duplicated().sum()
if duplicate_count == 0:
    print(Fore.GREEN + Style.BRIGHT + "No duplicate entries found in the dataset.")
else:
    print(Fore.GREEN + Style.BRIGHT + "Number of duplicate entries found: {duplicate_count}")
```

```python

# 4.Basic Details of the Data-Set

print("The shape =",data.shape)

# 4.1 Show information about the dataset
num_rows, num_cols = data.shape
num_features = num_cols - 1
num_data = num_rows * num_cols
print(f"Number of Rows: {num_rows}")
print(f"Number of Columns: {num_cols}")
print(f"Number of Features: {num_features}")
print(f"Number of All Data: {num_data}")
print("-----------------------------------------------------------------")

# 4.2 Get basic information
print(data.info())
print("-----------------------------------------------------------------")

# 4.3 Descriptive Statistics of Numeric Variables
data.describe().T.round(2)
```

```python

# 5.Data Plotting

# 5.1 Bar Graph
sns.catplot(data=data, x='quality', kind='count')
plt.title('Distribution of Wine Quality')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
print("--------------------------------------------------------------------------------------------------------------------------------------------")

# 5.2 Histograms
data.hist(figsize=(12, 8))
plt.tight_layout()
plt.show()
print("--------------------------------------------------------------------------------------------------------------------------------------------")

# 5.3 Pie Chart
data.quality.value_counts().plot.pie(autopct='%1.1f%%', shadow=True, figsize=(10, 10))
plt.title('Quality Graph')
plt.legend()
plt.show()
print("--------------------------------------------------------------------------------------------------------------------------------------------")

# 5.4 Correlation Matrix using Heatmap
correlation = data.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation, annot=True)
plt.title("Correlation Matrix")
plt.show()
```
